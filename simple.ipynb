```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("quikr_car.csv")
data
```

```python
data.info()
```

```python
data_c=data.copy()
```

```python
data_c=data_c[data_c["year"].str.isnumeric()]
```

```python
data_c["year"]=data_c["year"].astype(int)
```

```python
data_c=data_c[data_c["Price"]!="Ask For Price"]
data_c
```

```python
data_c["Price"]=data_c["Price"].str.replace(",","").astype(int)
```

```python
data_c["Price"]=data_c["Price"].astype(int)
```

```python
data_c["kms_driven"]=data_c["kms_driven"].str.split().str.get(0).str.replace(",","")
```

```python
data_c=data_c[data_c["kms_driven"].str.isnumeric()]
```

```python
data_c
```

```python
data_c["name"]=data_c["name"].str.split().str.slice(0,3).str.join(" ")
```

```python
cdata=data_c.reset_index(drop=True)
```

```python
cdata.to_csv("cleaned_data.csv")
```
